```python
from pandas import DataFrame
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
size = 10
```

```python
random.randint(5,20)
```

```python
idx = pd.date_range('2020-01-01', periods=size, freq='H')
fitness1 = [1/num**2 for num in range(1, size + 1) ] 
fitness2 = [5/num for num in range(1, size + 1) ] 
random_numbers= [random.randint(5,20) for num in range(1, size + 1) ]
```

```python
ts = pd.Series(range(len(idx)), index=idx)
```

```python
df1 = pd.DataFrame({'time': idx, 'fitness': fitness1, 'population': random_numbers})
df2 = pd.DataFrame({'time': idx, 'fitness': fitness2})
```

```python
ax = df1.plot(y = 'fitness', label="exp 1")
ax.set(xlabel="time", ylabel="fitness")
df2.plot( y = 'fitness', label="exp 2", ax=ax)
```
