```python
from pandas import DataFrame
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
import itertools
```

```python
size = 10
```

```python
idx = pd.date_range('2020-01-01', periods=size, freq='H')
fitness1 = [1/num**2 for num in range(1, size + 1) ] 
fitness2 = [5/num for num in range(1, size + 1) ] 
random_numbers= [random.randint(5,20) for num in range(1, size + 1) ]
```

```python
ts = pd.Series(range(len(idx)), index=idx)
```

```python
df1 = pd.DataFrame({'time': idx, 'fitness': fitness1, 'population': random_numbers})
df2 = pd.DataFrame({'time': idx, 'fitness': fitness2})
```

```python
ax = df1.plot(y = 'fitness', label="exp 1")
ax.set(xlabel="time", ylabel="fitness")
df2.plot( y = 'fitness', label="exp 2", ax=ax)
```

```python
ax = df1.plot(y=['population', 'fitness'])
ax.set(xlabel="time")
```

```python
number_of_repeatitions = 5
s = pd.Series(range(0, size)).repeat(number_of_repeatitions)
fitness3= [100-num for num in range(1, (size * number_of_repeatitions) + 1)]
```

```python
df3 = pd.DataFrame({'time': s, 'fitness': fitness3})
```

```python
df_processed = pd.DataFrame({'fitness': df3.groupby('time')['fitness'].mean()
                             , 'std': df3.groupby('time')['fitness'].std()})
```

```python
ax = df_processed.plot(y='fitness', yerr='std')
ax.set(xlabel="time", ylabel="fitness")
```
